```python
# Insurance Claims Workflow Automation

#This notebook demonstrates a simple rule -based claims atriage model used to simulate automatation oppotunities in insuance claim opeations.

#STEP 1 — Create Dummy Dataset

import pandas as pd
import numpy as np

# Set random seed for consistency
np.random.seed(42)

# Number of claims
num_claims = 500

# Generate dummy data
data = {
    "Claim_ID": range(1, num_claims + 1),
    "Claim_Amount": np.random.randint(200, 10000, num_claims),
    "Policy_Tenure_Months": np.random.randint(1, 60, num_claims),
    "Previous_Claims": np.random.randint(0, 5, num_claims),
    "Fraud_Flag": np.random.choice([0, 1], num_claims, p=[0.9, 0.1])
}

df = pd.DataFrame(data)

def risk_scoring(row):
    if (
        row["Claim_Amount"] < 1000 and
        row["Policy_Tenure_Months"] > 12 and
        row["Previous_Claims"] == 0 and
        row["Fraud_Flag"] == 0
    ):
        return "Low Risk"

    elif row["Fraud_Flag"] == 1 or row["Previous_Claims"] >= 3:
        return "High Risk"

    else:
        return "Medium Risk"

df["Risk_Category"] = df.apply(risk_scoring, axis=1)

# Calculate percentages
risk_counts = df["Risk_Category"].value_counts(normalize=True) *100

low_risk_percentage = risk_counts.get("Low Risk", 0)
medium_risk_percentage = risk_counts.get("Medium Risk", 0)
high_risk_percentage = risk_counts.get("High Risk", 0)

print(f"Low Risk (Auto Approved): {low_risk_percentage:.2f}%")
print(f"Medium Risk (Manual Review): {medium_risk_percentage:.2f}%")
print(f"High Risk (Escalation): {high_risk_percentage:.2f}%")

manual_cost_per_claim = 15
total_claims = 500

manual_processing_cost = total_claims * manual_cost_per_claim

auto_approved_claims = (low_risk_percentage / 100) * total_claims
manual_claims_after = total_claims - auto_approved_claims

new_processing_cost = manual_claims_after * manual_cost_per_claim

cost_savings = manual_processing_cost - new_processing_cost

print(f"Estimated Daily Cost Savings: £{cost_savings:.2f}")

```
